```python
# Sklearn Libraries
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

!pip install xgboost
# Regression Models
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
# Step 4: Train-Test Split
X = df.drop("Price", axis=1)  # Replace "target" with your actual target column
Y = output["Price"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
# Step 5: Define a dictionary of models
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "Random Forest": RandomForestRegressor(),
    "Gradient Boosting": GradientBoostingRegressor(),
    "Support Vector Regressor": SVR(),
    "XGBoost": XGBRegressor()
}
```

```python
# Step 6: Evaluate each model using Cross-Validation
model_results = {}

for name, model in models.items():
    cv_scores = cross_val_score(model, X_train, Y_train, cv=5, scoring="r2")
    model_results[name] = np.mean(cv_scores)
    print(f"{name} - Mean R² Score: {np.mean(cv_scores):.4f}")
```

```python
# Step 7: Select the Best Model
best_model_name = max(model_results, key=model_results.get)
best_model = models[best_model_name]

print(f"\nBest Model Based on R² Score: {best_model_name}")
```

```python
# Step 8: Hyperparameter Tuning for the Best Model
param_grid = {}

if best_model_name == "Random Forest":
    param_grid = {
        "n_estimators": [50, 100, 200],
        "max_depth": [5, 10, 20],
        "min_samples_split": [2, 5, 10]
    }
elif best_model_name == "Gradient Boosting":
    param_grid = {
        "n_estimators": [50, 100, 200],
        "learning_rate": [0.01, 0.1, 0.2],
        "max_depth": [3, 5, 10]
    }
elif best_model_name == "XGBoost":
    param_grid = {
        "n_estimators": [50, 100, 200],
        "learning_rate": [0.01, 0.1, 0.2],
        "max_depth": [3, 5, 10]
    }

if param_grid:
    grid_search = GridSearchCV(best_model, param_grid, cv=5, scoring="r2", n_jobs=-1)
    grid_search.fit(X_train, Y_train)
    best_model = grid_search.best_estimator_
    print(f"\nBest Parameters for {best_model_name}: {grid_search.best_params_}")
```

```python
# Step 9: Evaluate the Best Model on Test Data
Y_pred = best_model.predict(X_test)
print(Y_pred)
```

```python
r2 = r2_score(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))

print("\nFinal Model Performance on Test Data:")
print(f"R² Score: {r2:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
```
